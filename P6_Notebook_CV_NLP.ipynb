Votre mission est de réaliser une première étude de faisabilité d'un moteur de classification d'articles basé sur une image et une description pour l'automatisation de l'attribution de la catégorie de l'article.

Vous analyserez le jeu de données en réalisant un prétraitement des images et des descriptions des produits, une réduction de dimension, puis un clustering. Les résultats du clustering seront présentés sous la forme d’une représentation en deux dimensions à déterminer, qui ’illustrera le fait que les caractéristiques extraites permettent de regrouper des produits de même catégorie.

#### Pretreatment

**Following method will be graying images and resizing them  without distorsion (adding blank bands)**

```python
Sized_BW_dict = import_dict.copy()
hw_max = 512
for key, val in Sized_BW_dict.items():
    Sized_BW_dict[key] = ImageOps.grayscale(val)  # put image in grey scale
    Sized_BW_dict[key] = ImageOps.pad(Sized_BW_dict[key],
                                      # new squared image size
                                      (hw_max, hw_max),
                                      method=3,  # enlarging along one direction BICUBIC method
                                      color=255,  # filling with white
                                      # centering original image
                                      centering=(0.5, 0.5)
                                      )
```

#### Actual descriptors clustering

**Kmeans clustering to actually build image features**

- \# of cluster is fixed as sqrt of # descriptors

- Will fit 728 clusters

- Using minibatch to speed up computations

```python
from sklearn import cluster, metrics

k = int(round(np.sqrt(len(descriptors_all)),0))
print("Nombre de clusters estimés : ", k)
print("Création de",k, "clusters de descripteurs ...")

# Clustering
kmeans = cluster.MiniBatchKMeans(n_clusters=k, batch_size = 2048, init_size=3*k, random_state=13)
kmeans.fit(descriptors_all)
```

**Creation of histograms (filling features per pic)**

counting how many descriptors belong to clusters (= increasing feature values for each image)

```python
# Creation of histograms (features)

def build_histogram(kmeans, des, image_num):
    res = kmeans.predict(des)
    hist = np.zeros(len(kmeans.cluster_centers_))
    nb_des=len(des)
    if nb_des==0 : print("problème histogramme image  : ", image_num)
    for i in res:
        hist[i] += 1.0/nb_des
    return hist


# Creation of a matrix of histograms
hist_vectors=[]

for i, image_desc in enumerate(descriptors) :
    if i%100 == 0 : print(i)  
    hist = build_histogram(kmeans, image_desc, i) #calculates the histogram
    hist_vectors.append(hist) #histogram is the feature vector

im_features = np.asarray(hist_vectors)
```

### Dim Reduction : PCA + t-SNE

```python
from sklearn import manifold, decomposition

print("Dimensions dataset avant réduction PCA : ", im_features.shape)
pca = decomposition.PCA(n_components=0.99)
feat_pca= pca.fit_transform(im_features)
print("Dimensions dataset après réduction PCA : ", feat_pca.shape)
```

#### t-SNE

- Chi2 test show that they're not independant (*p-value* ~ 0), well that's fair enough...
- *Cramer coef* = 0.23 suggesting a substential association between the two

```python
from scipy import stats


def cramers_corrected_stat(x, y):
    """ calculate Cramers V statistic for categorial-categorial association.
        uses correction from Bergsma and Wicher, 
        Journal of the Korean Statistical Society 42 (2013): 323-328
    """
    result = -1
    conf_matrix = pd.crosstab(x, y)

    if conf_matrix.shape[0] == 2:
        correct = False
    else:
        correct = True

    chi2 = stats.chi2_contingency(conf_matrix, correction=correct)[0]

    n = sum(conf_matrix.sum())
    phi2 = chi2/n
    r, k = conf_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    result = np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))
    return round(result, 6)


labels_true = labels
labels_kmeans = cls.labels_
crosstab = pd.crosstab(labels, cls.labels_)

phi = cramers_corrected_stat(labels,cls.labels_)
print(f"Cramer's is {phi:.2}")

chi2, p, dof, ex = stats.chi2_contingency(crosstab, correction=False)
print(f"p-value is {p:.2} and chi2 = {chi2:.3}")
```

## Stem-Tokenizing, BoW, TF-IDF

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from nltk.stem.snowball import EnglishStemmer
```

### Defining variable, methods and pipeline

Scikit learn is very convenient because:

- I start to know fairly well the library
- the use of the pipeline is cool
- it's builtin tokenizer (CountVectorizer) already apply lower case, stop words and word frequency tresholding

However, no stemmer is built-in sklearn so we call nltk lib and define a new tokenizer allying the best of the two libs.

```python
X = df.description.to_numpy()

stemmer = EnglishStemmer()


class StemmedCountVectorizer(CountVectorizer):
    def build_analyzer(self):
        analyzer = super(StemmedCountVectorizer, self).build_analyzer()
        return lambda doc: ([stemmer.stem(w) for w in analyzer(doc)])


stemTokenizer = StemmedCountVectorizer(analyzer='word',
                                       max_features=1000,
                                       stop_words='english',
                                       max_df=0.20,
                                       min_df=3,
                                       dtype=np.int32
                                       )

pipe = Pipeline([('stemTokenizer', stemTokenizer),
                 ('tfidf', TfidfTransformer())
                 ]
                ).fit(X) # Note that this is very important to fit over the whole X set
```
